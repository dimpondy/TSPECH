#Représentation en Histogrammes de valeurs expérimentales

OBJECTIF:
*     Entrez vos données expérimentales
*     Modifier les lignes pour déteminer la moyenne et l'écart type

```
import matplotlib.pyplot as plt
import statistics

data = [1,2,2,2,3,3,3,4,4,5]
# On utlisera les modules sum() pour la somme et len() pour le nombre de données
moyenne=sum(data)/len(data)
plt.hist(data)
plt.title('Histogramme de données expérimentales', fontsize=12,c="red")
plt.xlabel('Valeurs')
plt.ylabel('Echantillons')
plt.axis([1, 5, 0, 5])
plt.savefig("plot_simple_histogramme_matplotlib_01.png")
plt.grid(True)
plt.show()
print('La moyenne est de ',moyenne)
#Alternative avec le module mean de la Bibiothèque Statistic
statistics.mean(data)
print('la pvariance est ' )
statistics.pvariance(data)
```

```
import matplotlib.pyplot as plt
import statistics

data = [1,2,2,2,3,3,3,4,4,5]

plt.hist(data)
plt.title('Histogramme de données expérimentales', fontsize=12,c="red")
plt.xlabel('Valeurs')
plt.ylabel('Echantillons')
plt.axis([1, 5, 0, 5])
plt.savefig("plot_simple_histogramme_matplotlib_01.png")
plt.grid(True)
plt.show()
print('La moyenne est de ',moyenne)

#pvariance avec le module de la Bibiothèque Statistic
#pvariance petit indique des données proches de la moyenne
statistics.pvariance(data)
```

```
import matplotlib.pyplot as plt
import random

# 1000 tirages entre 0 et 150
x = [random.randint(0,150) for i in range(1000)]
n, bins, patches = plt.hist(x, 70, normed=1 ,facecolor='r', alpha=1)
plt.xlabel('Mise')
plt.ylabel(u'Probabilite')
plt.axis([0, 150, 0, 0.02])
plt.grid(True)
plt.show()
```
